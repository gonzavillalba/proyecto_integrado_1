```python
# Importar libreras necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos desde GitHub
# Reemplazar con tu URL de GitHub raw
url = 'https://raw.githubusercontent.com/gonzavillalba/proyecto_integrado_1/refs/heads/main/ventas_tecnologia.csv?token=GHSAT0AAAAAAC2RLB77ACTRQYXO4HX3PROSZZXTHJA'
df = pd.read_csv(url)

# Crear grfico de lneas para todas las categoras
plt.figure(figsize=(15, 8))
for columna in df.columns[1:]:  # Excluimos la columna 'Mes'
    plt.plot(df['Mes'], df[columna], marker='o', label=columna)

plt.title('Ventas de Productos Tecnolgicos 2023')
plt.xlabel('Mes')
plt.ylabel('Unidades Vendidas')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
